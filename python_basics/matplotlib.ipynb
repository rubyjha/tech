# Matplotlib

Matplotlib is a ploting library and easy to get started for simple plots.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
## Simple Examples

x=np.arange(0,10)
y=np.arange(11,21)

a=np.arange(40,50)
b=np.arange(50,60)
```

```python
##plotting using matplotlib 

##plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'g*',linestyle='dashed', markersize=12)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('2D Plot')
```

```python
## Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*')

plt.subplot(2,2,3)
plt.plot(x,y,'b*')

plt.subplot(2,2,4)
plt.plot(x,y,'g*')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title('2D')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
```

```python

## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram")
```

```python
#Box Plot using Matplotlib
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
```

# Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```

```python
labels='a','b','c','d'
sizes=[1,2,3,5]
colors=['red','green','blue','lightskyblue']
explode=(0,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%', shadow=False)
plt.axis('equal')
```
